```python
from functools import*
l=[1,2,3,4]
x=list(map(lambda i:i*i,l))
print(x)
x1=list(filter(lambda i:i%2==1,l))
print(x1)
x3=reduce(lambda i,y:i*y,l)
print(x3)
```
