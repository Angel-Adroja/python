```python
l=[]
odd=0
even=0
for i in range(5):
    mark=input("enter mark student")
    l.append(mark)
print(l)
l.sort()
print(l)
print("max",l[-1])
print("min",l[0])
print("second max: ",l[-2])
print("second min: ",l[0])
for i in l:
    if i%2==0:
        even=even+l[i]
    else:
        odd=odd+l[i]
print("odd",odd)
print("even",even)
```
